<h4>1. Rendelések</h4>
<p>Az adatok a <i>orders.tsv</i> szöveg fájlban vannak, ami tabulárrral elválasztott szöveges állomány.<br/><b>a)</b> Olvassa be az adatokat egy dataframe-be. Ne felejtse beállítani, hogy az elválasztó karakter most a '\t'</p>

```python
import pandas as pd

#a feladat (2 pont)
df = pd.read_csv('orders.tsv', sep='\t')
dfcopy = df.copy()

df.head()
```

<p><b>b)</b> ellenőrizze a 'item_price' oszlop típusát. Alakítsa át, hogy szám legyen. <br/><i>(Segítség: 1)levágjuk a szöveg elejéről a dollár jelet, vagy 2) a dollár jelet egy üressel helyettesítjük)</i></p> 

```python
#b feladat (1 pont)
print(type(df['item_price'][0]))

df['item_price'] = df['item_price'].str.replace('$', '', regex=False)
df['item_price'] = pd.to_numeric(df['item_price'])

print(type(df['item_price'][0]))

df.head()
```

<p><b>c)</b> Számolja ki a rendelések összegét egy 'Sum' oszlopba (az ár egységárat jelent, és quantity oszlop adja meg, hogy hány darabot rendeltek az adott ételből)</p>

```python
#c feladat (1 pont)
df['Sum'] = df['quantity'] * df['item_price']

df.head()
```

<p><b>d)</b> Számítsa ki az átlagos rendelés értékét (az összes rendelés értékét kell osztani a rendelések számával)</p>

```python
#d feladat (1 pont)
df['Sum'].sum()/len(df['order_id'])
```

<p><b>e)</b> Írassuk ki, hogy az adott ételből hányat adtak el. (<i>Segítség: Az adott item_name csoporthoz összesen mennyit adtak el</i>)<br/>
<b>f)</b> Írassuk ki a rendelt mennyiség szerint rendezve.</p> 

```python
#e feladat (1 pont)
print(df.groupby(['item_name'])['quantity'].sum())

# Output exceeds the size limit. Open the full output data in a text editor!

#f feladat (1 pont)
print(df.groupby(['item_name'])['quantity'].sum().sort_values())
```

<p><b>g)</b> Az eredeti adatok közül gyűjtsük ki a dobozos üdítő italok adatait. (<i>Azaz "Canned Soda", vagy "Canned Soft Drink"</i>).</p>

```python
#g feladat (1 pont)
dfcopy.loc[(dfcopy['item_name'] == 'Canned Soda') | (dfcopy['item_name'] == 'Canned Soft Drink')]
```

<p><b>h)</b> Készítsünk egy listát, hogy az egyes üdítőkből összesen mennyit rendeltek.</p>

```python
#h feladat (1 pont)
dfcopy.loc[(df['item_name'] == 'Canned Soda') | (dfcopy['item_name'] == 'Canned Soft Drink')].groupby('item_name')['quantity'].sum()
```

<h4>2. Senzorok</h4>

<p>Az <a href="http://web.uni-corvinus.hu/~fszabina/data/adls_b.txt">adls_b.txt</a> és az <a href="http://web.uni-corvinus.hu/~fszabina/data/sensors_b.txt">sensors_b.txt</a> egy kísérleti alany szokásos otthoni tevékenységeiről tartalmaz adatokat. Az adls_b.txt azt adja meg, hogy az alany mely időszakban milyen jellegű tevékenységet végzett, a sensors_b.txt pedig a házban elhelyezett szenzorok mérési adatait tartalmazza.</p>
<p>Készítsen Python nyelvű programot, amely <br/>
<b>(a)</b> a tevékenységre fordított összidő alapján csökkenő sorrendben kiírja az alany tevékenységeit! <br/>(<i>Segítség: - A tevékenységek az Activity oszlopban találhatók.<br/>
 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - Az első két oszlopot dátumra kell konvertálni</i>)<br/>
<b>(b)</b> meghatározza, hogy a fürdőszobában (Bathroom) mely időintervallumokban érzékeltek a szenzorok aktivitást a 2012.11.15-ig terjedő időszakban!</p>

```python
import pandas as pd

activities_df = pd.read_csv('adls_b.txt', sep=';')
print(activities_df.head())

sensors_df = pd.read_csv('sensors_b.txt', sep=';')
print(sensors_df.head())
```

```python
#a feladat (1 pont)
activities_df['Starttime'] = pd.to_datetime(activities_df['Starttime'])
activities_df['Endtime'] = pd.to_datetime(activities_df['Endtime'])

activities_df['Activetime'] = activities_df['Endtime']-activities_df['Starttime']

print(activities_df.groupby(['Activity'])['Activetime'].sum().sort_values(ascending=False))
```

```python
#b feladat (1 pont)
sensors_df['Starttime'] = pd.to_datetime(sensors_df['Starttime'])
sensors_df['Endtime'] = pd.to_datetime(sensors_df['Endtime'])

sensors_df.loc[(sensors_df['Place'] == 'Bathroom') & (sensors_df['Endtime']<'2012-11-16')]
```
